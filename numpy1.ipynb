```python
""" Create an array with all zeros"""
Zeros = np.zeros(
    shape=(5,)
)
print(Zeros)
```

```python
""" Create an array with all ones"""
Ones = np.ones(
    shape=(4,),
    dtype=int
)
print(Ones)
```

```python
""" Create an array with specific value"""
specific_value = np.full(
    shape=(5,6),
    fill_value=3)
print(specific_value)
```

```python
""" Create an indentity matrix"""
Indentity_matrix = np.eye(
    3,
    dtype=int
)
print(Indentity_matrix)
```

```python
""" Create an empty array"""
My_empty_array = np.empty(
    shape=(4,)
)
print(My_empty_array)
print(My_empty_array.dtype)
```

```python
""" Type Casting"""
My_empty_array_int = My_empty_array.astype("int"
)
print(My_empty_array_int)
print(My_empty_array_int.dtype)
```

```python
""" Create a sequential array"""
My_array = np.arange(start=2, stop=11, step=2)
print(My_array)
```

```python
My_linespace_array = np.linspace(start=0, stop=15, num=3, dtype=int)
print(My_linespace_array)
```
