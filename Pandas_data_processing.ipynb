Handling Outliers

```python
Q1 = Health_data["Pulse"].quantile(0.25)
print(Q1)
Q3 = Health_data["Pulse"].quantile(0.75)
print(Q3)

IQR = Q3 - Q1
L = Q1 - 1.5 * IQR
R = Q3 + 1.5 * IQR
print(L , R)
```

```python
Non_Outlier_data = Health_data["Pulse"].between(L, R)
print(Non_Outlier_data)
```

```python
print(Health_data[~Non_Outlier_data])
```

```python
Health_data = Health_data[Non_Outlier_data]
Health_data.head()
```
