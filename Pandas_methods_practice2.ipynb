```python
import os
import pandas as pd
import numpy as np

Root_dir = os.getcwd()
Data_dir = os.path.join(Root_dir, "Data")
Dataset_path = os.path.join(Data_dir, "energy_sector_dataset.csv")
print(Root_dir)
print(Data_dir)
print(Dataset_path)
```

```python
energy_data = pd.read_csv(Dataset_path)
energy_data.head()
```

```python
energy_data.info()
```

```python
energy_data.isnull().sum()
```

```python
energy_data["date"] = energy_data["date"].astype("datetime64[ns]")
print(energy_data.head())
print(energy_data["date"].dtype)
```

```python
Q1 = energy_data["generation_mwh"].quantile(0.25)
print(Q1)
Q3 = energy_data["generation_mwh"].quantile(0.75)
print(Q3)
IQR = Q3 - Q1
L = Q1 - 1.5 * IQR
R = Q3 + 1.5 * IQR
print(L, R)
```

```python
Non_outlier_energy_data = energy_data["generation_mwh"].between(L, R)
outliered_energy_data = ~Non_outlier_energy_data
print(energy_data[outliered_energy_data])
```

```python
energy_data = energy_data[Non_outlier_energy_data]
energy_data.head()
```

```python
print(energy_data[outliered_energy_data])
```

```python
def normalization(col):
    Mean = col.mean()
    Std = col.std()
    return (col - Mean) / Std
energy_data["generation_mwh"] = energy_data["generation_mwh"].agg(normalization)
energy_data["capacity_mw"] = energy_data["capacity_mw"].agg(normalization)
energy_data["co2_emissions_tons"] = energy_data["co2_emissions_tons"].agg(normalization)
energy_data["energy_price_eur_mwh"] = energy_data["energy_price_eur_mwh"].agg(normalization)
energy_data.head()
```

```python
energy_data = energy_data.drop(columns=["plant_id"])
energy_data.head()
```

```python
sorted_energy_data = energy_data.sort_values(
    by = ["plant_type"],
    ascending= True
)
sorted_energy_data.head(10)
```
