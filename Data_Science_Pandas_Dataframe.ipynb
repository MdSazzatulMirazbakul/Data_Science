Sorting

```python
sorted_homelessness_df = homelessness_df.sort_values(
    by=["family_members"],
    ascending=[True]
)
sorted_homelessness_df.head(10)
```

```python
sorted_homelessness_df = homelessness_df.sort_values(
    by=["region", "family_members"],
    ascending=[True, False]
)
sorted_homelessness_df.head(10)
```

Feature Engineering

```python
homelessness_df.head()
print(homelessness_df.columns)
```

```python
New_homelessness_df = homelessness_df[["region", "individuals", "family_members"]]
New_homelessness_df.head()
```

```python
sorted_new_homelessness_df = New_homelessness_df.sort_values(
    by = ["region", "individuals"],
    ascending=[True, False]
)
sorted_new_homelessness_df.head(10)
```

Filtering

```python
mask = sorted_new_homelessness_df["region"] == "Mountain"
Filtered_df_mountain = sorted_new_homelessness_df[mask]
print(Filtered_df_mountain)
```

```python
homelessness_df_10k = sorted_new_homelessness_df[sorted_homelessness_df["individuals"] > 10000]
homelessness_df_10k.head(10)
```

```python
family_10k_mask = sorted_homelessness_df["family_members"] < 10000
region_pacific = sorted_homelessness_df["region"] == "Pacific"
New_filtered_df = sorted_homelessness_df[family_10k_mask & region_pacific]
New_filtered_df.head()
```

```python
new_df = sorted_homelessness_df.query(
    "region == 'Pacific' and individuals > 10000"
)
new_df.head()
```

```python
New_df = sorted_homelessness_df.query(
    "state == 'Washington' and individuals > 10000"
)
New_df.head()
```
