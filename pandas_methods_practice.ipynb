```python
import os
import pandas as pd
import numpy as np

root_dir = os.getcwd()
data_dir = os.path.join(root_dir, "Data")
dataset_path = os.path.join(data_dir, "sales_subset.csv")

print("root dir: ", root_dir)
print("data dir: ", data_dir)
print("dataset path: ", dataset_path)
```

```python
sales_data = pd.read_csv(dataset_path)
sales_data.head()
```

```python
sales_data.columns
```

```python
sales_data = sales_data.drop(columns = ["department"])
sales_data.head()
```

```python
sales_data.columns = ["Store", "Type", "Date", "Weekly Sales", "Holiday or NOt", "Temperature", "Fuel Price USD", "Unemployment"]
sales_data.head()
```

```python
sales_data.info()
```

```python
Avarage_weekly_sales = sales_data[["Weekly Sales", "Fuel Price USD"]].mean()
print(Avarage_weekly_sales)
```

```python
average_weekly_sales = np.mean(sales_data[["Weekly Sales"]])
print(average_weekly_sales)
```

```python
print("mean: ", sales_data["Weekly Sales"].mean())
print("median: ", sales_data["Weekly Sales"].median())
print("Max: ", sales_data["Date"].max())
print("Min: ", sales_data["Date"].min())
print("sum: ", sales_data["Weekly Sales"].sum())
```

Custom Method for Normalization or Standardization

```python
sales_data.head()
```

```python
def standardization(col):
    Mean = col.mean()
    Std = col.std()
    return (col - Mean) / Std
Normalized_Weekly_Sales_and_Temp = sales_data[["Weekly Sales", "Temperature"]].agg([standardization, np.cumsum])
Normalized_Weekly_Sales_and_Temp.head()
```

```python
Normalized_Weekly_Sales_and_Temp.max(), Normalized_Weekly_Sales_and_Temp.min()
```

```python
"""Group By"""
Sales_grouped_data = sales_data.groupby("Type")
Sales_grouped_data.head(10)
```

```python
Sales_statistics = sales_data.groupby("Type")["Weekly Sales"].agg([min, max, np.mean, np.median])
Sales_statistics.head()
```

```python
""" Pivot Table"""
Sales_Pivot = sales_data.pivot_table(
    index= "Type",
    values= "Weekly Sales",
    aggfunc=[min, max, np.mean, np.median]
)
Sales_Pivot
```

```python
Sales_pivot = sales_data.pivot_table(
    index= "Type",
    values= "Weekly Sales",
    columns= "Holiday or NOt",
    aggfunc= [np.mean]
)
Sales_pivot
```
