```python
Homelessness_data = pd.read_csv("D:\\Jupyter\\Data\\homelessness.csv")
# Homelessness_data.head()
Homelessness_data.tail(10)
```

```python
""""Feature Engineering"""
New_homelessness_data = Homelessness_data[["region", "state", "individuals", "family_members", "state_pop"]]
New_homelessness_data.head()
```

```python
"""Adding a New column"""
New_homelessness_data["All_Population"] = New_homelessness_data["individuals"] + New_homelessness_data["state_pop"]
New_homelessness_data.head(10)
```

```python
""""Sorting"""
sorted_homelessness_data = Homelessness_data.sort_values(
    by= ["region"],
    ascending= True
)
sorted_homelessness_data.head(10)
```

```python
sorted_homelessness_data_multi_col = New_homelessness_data.sort_values(
    by= ["region", "individuals"],
    ascending= [True, True]
)
sorted_homelessness_data_multi_col.head(10)
```

```python
""""Selecting Rows"""
mask = New_homelessness_data["region"] == "Mid-Atlantic"
New_homelessness_data_Mid_Atlantic = New_homelessness_data[mask]
New_homelessness_data_Mid_Atlantic
```

```python
mask_2 = New_homelessness_data["region"] == "Mountain"
New_homelessness_data_Mountain = New_homelessness_data[mask_2]
New_homelessness_data_Mountain
```

```python
mask_3 = New_homelessness_data["individuals"] > 10000
New_homelessness_data_less_10k = New_homelessness_data[mask_3]
New_homelessness_data_less_10k
```

```python
New_homelessness_data_less_10k_family = New_homelessness_data[New_homelessness_data["family_members"] < 10000]
New_homelessness_data_less_10k_family.head(10)
sorted_new_homelessness_data_less_10k_family = New_homelessness_data_less_10k_family.sort_values(
    by= ["family_members"],
    ascending= [True]
)
sorted_new_homelessness_data_less_10k_family
```

```python
"""Query"""
New_homelessness_data_sorted = New_homelessness_data.query(
    "region == 'Pacific' and individuals > 10000"
)
New_homelessness_data_sorted
```
