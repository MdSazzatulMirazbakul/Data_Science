```python
"""" Spliting dataset into training, testing and validation"""
data_set = np.random.randint(100, size=(50, 3))
print("Data set:\n ", data_set[:5])
```

```python
np.random.shuffle(data_set)
print(data_set[:5])
```

```python
""""Spliting dataset"""
index_1 = int(0.8 * len(data_set))
index_2 = int(0.9 * len(data_set))
training, testing, validation = np.split(
    data_set, [index_1, index_2]
)
print(len(training), len(testing), len(validation))
```
