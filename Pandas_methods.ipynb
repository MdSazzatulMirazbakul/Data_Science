Normalisation or standardization

```python
def normalize(col):
    mean = col.mean()
    std = col.std()
    return (col - mean) / std
normalized_sales_temp = Sales[["Weekly Sales", "Temperature C"]].agg([normalize, np.cumsum])
normalized_sales_temp
```

```python
"""Pandas methods for catagorical data"""
Sales[["Type"]].value_counts()
```

```python
"""Group By"""
Sales_groupby_type = Sales.groupby("Type")
Sales_groupby_type.head()
```

```python
Sales_groupby_type["Weekly Sales"].mean()
```

```python
sales_stats = Sales.groupby("Type")["Weekly Sales"].agg([max, min, np.mean, np.median])
sales_stats
```

```python
mean_sales_by_type = Sales.pivot_table(
    values= ["Weekly Sales"],
    index= "Type",
    aggfunc= [min, max, np.mean, np.median]
)
mean_sales_by_type
```

```python
mean_sales_by_type1 = Sales.pivot_table(
    values= ['Weekly Sales'],
    index= "Type",
    columns= ["Holiday"],
    aggfunc= [min, max, np.mean, np.median]
)
mean_sales_by_type1
```

```python
def standardization(col):
    Mean = col.mean()
    Std = col.std()
    return (col - Mean) / Std
Standardized_weekly_sales_and_temp = Sales[["Weekly Sales", "Temperature C"]].agg([standardization])
Standardized_weekly_sales_and_temp.head()
```
